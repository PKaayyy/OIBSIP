 Car price prediction using Python involves several key steps: collecting data from reliable sources, preprocessing it by cleaning and transforming features, and performing exploratory data analysis (EDA) to understand relationships within the data.


```python
#Importing Libraries
import pandas as pd
import numpy as np
from sklearn import linear_model

```

```python
#Reading CSV
df=pd.read_csv('/content/car data.csv')
df
```

```python
#Creating daTaframes
inputs=df.drop(['Car_Name','Owner','Selling_type'],axis='columns')
target=df.Selling_Price
target
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
Numerics=LabelEncoder()
```

```python
#New encoded columns
inputs['Fuel_Type_n']=Numerics.fit_transform(inputs['Fuel_Type'])
inputs['Transmission_n']=Numerics.fit_transform(inputs['Transmission'])
inputs
```

```python
#Dropping string columns
inputs_n=inputs.drop(['Fuel_Type','Transmission','Selling_Price',],axis='columns')
inputs_n
```

```python
#Linear regression
model=linear_model.LinearRegression()
```

```python
model.fit(inputs_n,target)
```

```python
#PREDICTION
pred=model.predict([[2009,	11.00,	87934,	2,	1]])
print(pred)
```
